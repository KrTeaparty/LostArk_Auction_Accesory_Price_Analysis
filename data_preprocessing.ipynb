```python
def extract_data(x):
    res = []
    js = eval(x.loc['Result'])
    x.loc['EquipType'] = js['Name'].split(' ')[-1]
    x.loc['GradeQuality'] = js['GradeQuality']
    x.loc['BuyPrice'] = js['AuctionInfo']['BuyPrice']
    x.loc['TradeAllowCount'] = js['AuctionInfo']['TradeAllowCount']

    options = js['Options']
    x.loc['FirstStatName'] = options[0]['OptionName']
    x.loc['FirstStatValue'] = options[0]['Value']

    if len(options) == 4:
        x.loc['SecondStatName'] = '-'
        x.loc['SecondStatValue'] = 0
    else:
        x.loc['SecondStatName'] = options[1]['OptionName']
        x.loc['SecondStatValue'] = options[1]['Value']

    x.loc['PenaltyEngraveName'] = options[-3]['OptionName']
    x.loc['PenaltyEngraveValue'] = options[-3]['Value']

    fev = options[-2]['Value']
    sev = options[-1]['Value']
    if fev < sev:
        x.loc['FirstEngraveName'] = options[-1]['OptionName']
        x.loc['FirstEngraveValue'] = options[-1]['Value']
        x.loc['SecondEngraveName'] = options[-2]['OptionName']
        x.loc['SecondEngraveValue'] = options[-2]['Value']
    else:
        x.loc['FirstEngraveName'] = options[-2]['OptionName']
        x.loc['FirstEngraveValue'] = options[-2]['Value']
        x.loc['SecondEngraveName'] = options[-1]['OptionName']
        x.loc['SecondEngraveValue'] = options[-1]['Value']

    isclassengrave = 0
    for option in options[-2:]:
        if option['ClassName']:
            isclassengrave = 1

    x.loc['ClassEngrave'] = isclassengrave
    return x
```

```python
df_result = df.apply(lambda x : extract_data(x), axis=1).drop('Result', axis=1)
```

```python
file_list = [f'auction_data/lostark_auction_data{i}.csv' for i in range(1, 16)]
```

```python
res_df = pd.DataFrame()

for file in file_list:
    df = pd.read_csv(file)
    res = df.apply(lambda x : extract_data(x), axis=1).drop('Result', axis=1)
    res_df = pd.concat([res_df, res], ignore_index=True)

res_df.to_csv('Preprocessed_Auction_Data.csv', index=False)
```
